### Handling Outliers ###

```python
#definisikan rumus mencari outliers
def lower_upper_limit(x):
    x75 = np.percentile(x, 75)
    x25 = np.percentile(x, 25)
    iqr = x75-x25
    lower_lim = x25 - (1.5 * iqr)
    upper_lim = x75 + (1.5 * iqr)
    return lower_lim, upper_lim

def outliers(x):
    batas_bawah, batas_atas = lower_upper_limit(x)
    return x[np.where(
    (x < batas_bawah) | 
    (x > batas_atas))]
```

```python
# Outliers GDP
print('Outliers GDP adalah: ', outliers(df2['GDPperkapita'].values))

# Outliers Kesehatan
print('\nOutliers Kesehatan adalah: ', outliers(df2['Kesehatan'].values))
```

```python
# BB BA GDP
bb, ba = lower_upper_limit(df2['GDPperkapita'])
print(bb, ba, sep=' ')

# BB BA Kesehatan
bb1, ba1 = lower_upper_limit(df2['Kesehatan'])
print(bb1, ba1, sep=' ')
```

### Scaling ###

```python
# Data yang tanpa outliers ada di dataframe bernama df6_normal
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

df7 = df6_normal[['GDPperkapita','Kesehatan']]
sc= StandardScaler()
df7_scaled = sc.fit_transform(df7.astype('float'))
df7_scaled
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 50).fit(df7_scaled)
labels = kmeans.labels_
```

```python
df8 = pd.DataFrame(data= df7_scaled, columns=['GDP', 'Kesehatan'])
df8['hasil_kmeans']= labels
df8
```

```python
k1_x1 = df8.GDP[df8.hasil_kmeans == 0]
k1_y1 = df8.Kesehatan[df8.hasil_kmeans == 0]

k2_x1 = df8.GDP[df8.hasil_kmeans == 1]
k2_y1 = df8.Kesehatan[df8.hasil_kmeans == 1]

k3_x1 = df8.GDP[df8.hasil_kmeans == 2]
k3_y1 = df8.Kesehatan[df8.hasil_kmeans == 2]

k4_x1 = df8.GDP[df8.hasil_kmeans == 3]
k4_y1 = df8.Kesehatan[df8.hasil_kmeans == 3]

plt.subplots(figsize=(15,8))
plt.scatter(k1_x1, k1_y1, c='red', s=100, edgecolors='green', linestyle='-')
plt.scatter(k2_x1, k2_y1, c='blue', s=100, edgecolors='green', linestyle='-')
plt.scatter(k3_x1, k3_y1, c='yellow', s=100, edgecolors='green', linestyle='-')
plt.scatter(k4_x1, k4_y1, c='purple', s=100, edgecolors='green', linestyle='-')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='white', s=500)
plt.xlabel('GDP')
plt.ylabel('Kesehatan')
plt.show()
```

```python
# Elbow_Method
wcss = []
for i in range(1,15):
    kmeans= KMeans(n_clusters= i, init='k-means++', random_state= 50)
    kmeans.fit(df7_scaled)
    wcss.append(kmeans.inertia_)

plt.subplots(figsize=(15,8))
plt.plot(range(1,15), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
